```python
import matplotlib.pyplot as plt

n, t = [], []
with open("test_result.txt") as f:
    for line in f:
        x, y = line[:-1].split(" ")
        n.append(int(x))
        t.append(float(y))
plt.plot(n, t)
plt.title("CPU Implementation Performance")
plt.xlabel("number of images")
plt.ylabel("time (s)")
```
